```r
rm(list=ls())
library(ggplot2)
```

## CDF of distance of differential peak to nearest differential gene

```r
dist_diff_peak_to_dist_gene=read.table("all.diffpeaks.promoters.closest.tsv",header=FALSE,sep='\t')
```

```r
head(dist_diff_peak_to_dist_gene)
```

```r

dist_diff_peak_to_dist_gene$V8[dist_diff_peak_to_dist_gene$V8 < 1]=1
dist_diff_peak_to_dist_gene$V8[dist_diff_peak_to_dist_gene$V8< -1]=1000000
```

```r
p1=ggplot()+
  stat_ecdf(data=dist_diff_peak_to_dist_gene,aes(x=dist_diff_peak_to_dist_gene$V8 ,color="Diff Peaks Pre Duplicate Removal"))+
  geom_vline(aes(xintercept=1000,color="1kb"))+
  geom_vline(aes(xintercept=5000,color="5kb"))+
  geom_vline(aes(xintercept=10000,color="10kb"))+
  geom_vline(aes(xintercept=50000,color="50kb"))+
  geom_vline(aes(xintercept=100000,color="100kb"))+
  geom_vline(aes(xintercept=500000,color="500kb"))+
  geom_vline(aes(xintercept=1000000,color="1MB"))+
  geom_vline(aes(xintercept=2000000,color="2MB"))+
  scale_x_continuous(trans='log10') +
  xlab("Distance from differential promoter peak (N=758) to nearest differential gene")+
  ylab("CDF")+
scale_color_manual(values=c('#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f','#ff7f00','#cab2d6','#6a3d9a','#ffff99','#b15928'))

p1
```
