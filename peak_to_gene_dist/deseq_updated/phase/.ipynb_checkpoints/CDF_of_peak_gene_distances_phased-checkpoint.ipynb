## CDF of distance of differential peak to nearest differential gene, phased by cell cycle

```r
earlyG1_up_dist=read.table("earlyG1.up.dist.bed",header=FALSE,sep='\t')
earlyG1_down_dist=read.table("earlyG1.down.dist.bed",header=FALSE,sep='\t')

lateG1_up_dist=read.table("lateG1.up.dist.bed",header=FALSE,sep='\t')
lateG1_down_dist=read.table("lateG1.down.dist.bed",header=FALSE,sep='\t')

SG2M_up_dist=read.table("SG2M.up.dist.bed",header=FALSE,sep='\t')
SG2M_down_dist=read.table("SG2M.down.dist.bed",header=FALSE,sep='\t')

head(earlyG1_up_dist)
```

```r
earlyG1_up_dist$V8[earlyG1_up_dist$V8 < 1]=1
earlyG1_up_dist$V8[earlyG1_up_dist$V8< -1]=1000000

earlyG1_down_dist$V8[earlyG1_down_dist$V8 < 1]=1
earlyG1_down_dist$V8[earlyG1_down_dist$V8< -1]=1000000

lateG1_up_dist$V8[lateG1_up_dist$V8 < 1]=1
lateG1_up_dist$V8[lateG1_up_dist$V8< -1]=1000000

lateG1_down_dist$V8[lateG1_down_dist$V8 < 1]=1
lateG1_down_dist$V8[lateG1_down_dist$V8< -1]=1000000

SG2M_up_dist$V8[SG2M_up_dist$V8 < 1]=1
SG2M_up_dist$V8[SG2M_up_dist$V8< -1]=1000000

SG2M_down_dist$V8[SG2M_down_dist$V8 < 1]=1
SG2M_down_dist$V8[SG2M_down_dist$V8< -1]=1000000
```

```r
head(earlyG1_up_dist)
```

```r
p1=ggplot()+
  stat_ecdf(data=earlyG1_up_dist,aes(x=earlyG1_up_dist$V8 ,color="earlyG1_Up"))+
  stat_ecdf(data=earlyG1_down_dist,aes(x=earlyG1_down_dist$V8 ,color="earlyG1_Down"))+
  stat_ecdf(data=lateG1_up_dist,aes(x=lateG1_up_dist$V8 ,color="lateG1_Up"))+
  stat_ecdf(data=lateG1_down_dist,aes(x=lateG1_down_dist$V8 ,color="lateG1_Down"))+
  stat_ecdf(data=SG2M_up_dist,aes(x=SG2M_up_dist$V8 ,color="SG2M_Up"))+
  stat_ecdf(data=SG2M_down_dist,aes(x=SG2M_down_dist$V8 ,color="SG2M_Down"))+
  geom_vline(aes(xintercept=1000,color="thresh"))+
  geom_vline(aes(xintercept=5000,color="thresh"))+
  geom_vline(aes(xintercept=10000,color="thresh"))+
  geom_vline(aes(xintercept=50000,color="thresh"))+
  geom_vline(aes(xintercept=100000,color="thresh"))+
  geom_vline(aes(xintercept=500000,color="thresh"))+
  geom_vline(aes(xintercept=1000000,color="thresh"))+
  geom_vline(aes(xintercept=2000000,color="thresh"))+
  scale_x_continuous(trans='log10') +
  xlab("Phased Distance from differential peak (N=3245) to nearest differential gene")+
  ylab("CDF")+
scale_color_manual(values=c('#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#000000'))

p1
```
