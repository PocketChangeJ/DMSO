## CDF of distance of differential gene to nearest differential peak

```r
dist_diff_gene_to_diff_peak=read.table("genes.closest.filtered.txt",header=FALSE,sep='\t')
dist_diff_gene_to_all_peak=read.table("genes.closest.allcalledpeaks.filtered.bed",header=FALSE,sep='\t')
names(dist_diff_gene_to_diff_peak)=c("Gene","Dist")
names(dist_diff_gene_to_all_peak)=c("Gene","Dist")

dist_diff_gene_to_diff_peak$Dist[dist_diff_gene_to_diff_peak$Dist<1]=1
dist_diff_gene_to_diff_peak$Dist[dist_diff_gene_to_diff_peak$Dist< -1]=1000000


dist_diff_gene_to_all_peak$Dist[dist_diff_gene_to_all_peak$Dist<1]=1
dist_diff_gene_to_all_peak$Dist[dist_diff_gene_to_all_peak$Dist< -1]=1000000
```

```r
p1=ggplot()+
  stat_ecdf(data=dist_diff_gene_to_diff_peak,aes(x=dist_diff_gene_to_diff_peak$Dist ,color="Differential Peaks"))+
  stat_ecdf(data=dist_diff_gene_to_all_peak,aes(x=dist_diff_gene_to_all_peak$Dist,color="All Called Peaks"))+
  geom_vline(aes(xintercept=1000,color="1kb"))+
  geom_vline(aes(xintercept=5000,color="5kb"))+
  geom_vline(aes(xintercept=10000,color="10kb"))+
  geom_vline(aes(xintercept=50000,color="50kb"))+
  geom_vline(aes(xintercept=100000,color="100kb"))+
  geom_vline(aes(xintercept=500000,color="500kb"))+
  geom_vline(aes(xintercept=1000000,color="1MB"))+
  geom_vline(aes(xintercept=2000000,color="2MB"))+
  scale_x_continuous(trans='log10') +
  xlab("Distance from promoter of differential gene (N=884) to nearest (differential/called) peak")+
  ylab("CDF")+
scale_color_manual(values=c('#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f','#ff7f00','#cab2d6','#6a3d9a','#ffff99','#b15928'))

p1
```

```r
dist_diff_peak_to_dist_gene=read.table("peaks.closest.bed",header=FALSE,sep='\t')
dist_diff_peak_dedup_to_dist_gene=read.table("dupfilt/peak.dist.bed",header=FALSE,sep='\t')
```

## Differential genes: Distance to nearest differential peak vs adjusted p-value of gene 

```r
pval_genes=read.table("get_pvalues/gene.dist.to.pval.tsv",header=TRUE,sep='\t')
pval_genes$GeneDist[pval_genes$GeneDist<1]=1
head(pval_genes)
```

```r
print(ggplot(data=pval_genes,
            aes(x=-log10(pval_genes$GenePval),
               y=log10(pval_genes$GeneDist)))+
     geom_point(alpha=0.1)+
     xlab("-log10(p-adjusted)")+
     ylab("log10(Distance to nearest differential peak)")+
      ggtitle("Differential Genes: p-adj vs distance to differential peak"))
```

## Differential peaks: Distance to nearest differential gene vs adjusted p-value of peak 

```r
pval_peaks=read.table("get_pvalues/peak.dist.to.pval.tsv",header=TRUE,sep='\t')
pval_peaks$PeakDist[pval_peaks$PeakDist<1]=1
head(pval_peaks)
```

```r
print(ggplot(data=pval_peaks,
            aes(x=-log10(pval_peaks$PeakPval),
               y=log10(pval_peaks$PeakDist)))+
     geom_point(alpha=0.1)+
     xlab("-log10(p-adjusted)")+
     ylab("log10(Distance to nearest differential gene)")+
      ggtitle("Differential Peaks: p-adj vs distance to differential gene"))
```
